```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
token = Tokenizer()
```

```python
token.fit_on_texts([faqs])
```

```python
len(token.word_index)
```

```python
token.word_index
```

```python
input_sequences = []
for sentence  in faqs.split('\n'):
  # print(sentence)
  # print(token.texts_to_sequences([sentence])[0])
  token_sentence=token.texts_to_sequences([sentence])[0]
  for i in range(1,len(token_sentence)):
    input_sequences.append( token_sentence[:i+1])
```

```python
input_sequences
```

```python
max_len =max([len(x) for x in input_sequences])
print(max_len)
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding='pre')
```

```python
# print(padded_input_sequences)
```

```python
X = padded_input_sequences[:,:-1]
```

```python
Y = padded_input_sequences[:,-1]
```

```python
X.shape
```

```python
Y.shape
```

```python
from tensorflow.keras.utils import to_categorical
Y = to_categorical(Y,num_classes=408)
```

```python
Y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense
```

```python
model = Sequential()
model.add(Embedding(408,100,input_length=30))
model.add(LSTM(150))
model.add(Dense(408, activation="softmax"))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X,Y,epochs=100)
```
